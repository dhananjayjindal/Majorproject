```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('500112.csv')
```

```python
df
```

```python
# Dropping extra columns
df.drop(['year'], axis=1, inplace=True)
axis = df.shape
```

```python
no_of_stocks = axis[1]
columns = df.columns
```

```python
# Step 1 calculate returns (stockT2 - stockT1)/stockT1
returns = pd.DataFrame(index=range(axis[0]), columns=range(axis[1]))
for idx, col in enumerate(columns):
    for i in range(1, len(df[col])):
        s1 = df[col][i - 1]
        s2 = df[col][i]
        return_value = (s2 - s1) / s1
        returns[idx][i] = return_value

# returns = returns*100
returns.drop([0], axis=0, inplace=True)
```

```python
returns
```

```python
# Step 2 Calculation of Σ Covariance matrix between stocks from csv
Σ = returns.cov()
```

```python
# Step 3 calcuating quadtric programming equation or objective function
```

```python
# Step 3 calcualting geomean of returns constraint 1
returns_for_geomean = returns + 1
returns_for_geomean[0]
geo_mean = [
    (returns_for_geomean[i].prod() ** (1 / len(returns_for_geomean[i])) - 1)
    for i in returns_for_geomean
]
geo_mean = pd.DataFrame(geo_mean)

Σ.to_csv('covariance_matrix.csv', index=False)
geo_mean.to_csv('geomean.csv', index=False) 
```

```python
# Step 5 mean and of returns
µ = returns.mean()
µ.to_csv('mean_of_returns.csv', index=False) 
```
